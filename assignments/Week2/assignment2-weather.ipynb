### Create a jupyter notebook called assignment2-weather.ipynb that has a nice plot of the temperature over time ( "dryBulbTemperature_Celsius" ). 

```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates

weatherdata = pd.read_csv('C:/Users/gusgo/PFDA/PFDA/assignments/week2/weatherreadings1.csv', header=0)  # read the data from the csv file

weatherdata
```

Inspect Data

```python
weatherdata.iloc[0]
```

```python
weatherdata['dryBulbTemperature_Celsius'].value_counts()
```

```python
weatherdata.describe()
```

```python
temp_drycelsius = weatherdata[['reportStartDateTime', 'dryBulbTemperature_Celsius']]
temp_drycelsius
```

Line Plot

```python
plt.figure(figsize=(10, 6))
plt.plot(temp_drycelsius['reportStartDateTime'], temp_drycelsius['dryBulbTemperature_Celsius'], marker='o')
plt.title('Dry Bulb Temperature Over Time')
plt.xlabel('DateTime')
plt.ylabel('Dry Bulb Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(temp_drycelsius['reportStartDateTime'], temp_drycelsius['dryBulbTemperature_Celsius'], marker='o')
plt.title('Dry Bulb Temperature Over Time')
plt.xlabel('DateTime')
plt.ylabel('Dry Bulb Temperature (°C)')
plt.grid(True)

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))  # Set major ticks to every day
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # Format major ticks as 'YYYY-MM-DD'
ax.xaxis.set_minor_locator(mdates.HourLocator(interval=6))  # Set minor ticks to every 6 hours

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

Scatter Plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(temp_drycelsius['reportStartDateTime'], temp_drycelsius['dryBulbTemperature_Celsius'])
plt.title('Dry Bulb Temperature Over Time')
plt.xlabel('DateTime')
plt.ylabel('Dry Bulb Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(temp_drycelsius['reportStartDateTime'], temp_drycelsius['dryBulbTemperature_Celsius'], marker='o')
plt.title('Dry Bulb Temperature Over Time')
plt.xlabel('DateTime')
plt.ylabel('Dry Bulb Temperature (°C)')
plt.grid(True)

# Set major and minor ticks format and interval
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))  # Set major ticks to every day
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # Format major ticks as 'YYYY-MM-DD'
ax.xaxis.set_minor_locator(mdates.HourLocator(interval=6))  # Set minor ticks to every 6 hours

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(temp_drycelsius['dryBulbTemperature_Celsius'], bins=20, edgecolor='black')
plt.title('Distribution of Dry Bulb Temperature')
plt.xlabel('Dry Bulb Temperature (°C)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(temp_drycelsius['dryBulbTemperature_Celsius'], vert=False)
plt.title('Box Plot of Dry Bulb Temperature')
plt.xlabel('Dry Bulb Temperature (°C)')
plt.grid(True)
plt.show()
```
